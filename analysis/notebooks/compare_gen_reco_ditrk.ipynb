```python
def getRSME(reco, gen):
    err = ((reco-gen)/gen)
    #return np.sum(np.abs(err))/len(err)
    return np.sqrt(np.sum(err**2)/len(err))
```

```python
canvas = ROOT.TCanvas("canvas", "canvas", 1800, 600)
canvas.Divide(3, 1)

nbins, xlow, xhigh, = 100, -20, 20 #pt
#nbins, xlow, xhigh, = 200, -0.004, 0.004 #eta
#nbins, xlow, xhigh, = 200, -0.004, 0.004 #phi
nbins, xlow, xhigh, = 200, -0.2, 0.2 #mass
var, part = "mass", "ditrk"


h_old = dfOld.Histo1D(("1Dhist", "Diff_{}_{} OLD".format(var, part), nbins, xlow, xhigh), "Diff_{}_{}".format(part, var))
h_mass = dfMass.Histo1D(("1Dhist", "Diff_{}_{} MASS".format(var, part), nbins, xlow, xhigh), "Diff_{}_{}".format(part, var))

h_old.SetLineColor(ROOT.kBlue)
h_mass.SetLineColor(ROOT.kRed)
canvas.cd(1)
h_old.Draw("hist")
canvas.cd(2)
h_mass.Draw("hist")

h_old.Scale(1/h_old.GetEntries())
h_mass.Scale(1/h_mass.GetEntries())

canvas.cd(3)
stack = ROOT.THStack("stack", "reco-gen")
stack.Add(h_old.GetValue())
stack.Add(h_mass.GetValue())
stack.Draw("hist nostack")

#stack.GetXaxis().SetRangeUser(100, 150)

legend = ROOT.TLegend(0.60, 0.65, 0.8999, 0.89)
legend.SetMargin(0.17)
legend.SetBorderSize(0)
legend.SetTextSize(0.04)
legend.AddEntry(h_old.GetValue(), "OLD", "l")
legend.AddEntry(h_mass.GetValue(), "NMass", "l")
legend.Draw()

canvas.SetGrid()
canvas.Draw()
```
