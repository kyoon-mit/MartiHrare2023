```python
cols = ["goodMeson_ditrk_pt_GEN", "goodMeson_ditrk_pt", "goodMeson_ditrk_mass_GEN", "goodMeson_Nphotons"]
x = df.AsNumpy(columns=cols)
pddf = pd.DataFrame(x)
pddf["goodMeson_Nphotons"] = pddf["goodMeson_Nphotons"].apply(lambda x: x[0])

display(pddf[pddf["goodMeson_Nphotons"] == 1])
display(pddf[pddf["goodMeson_Nphotons"] == 2])
display(pddf[pddf["goodMeson_Nphotons"] == 3])
display(pddf[pddf["goodMeson_Nphotons"] > 3])
```

```python
canvas = ROOT.TCanvas("canvas", "canvas", 1800, 600)
canvas.Divide(3, 1)

nbins, xlow, xhigh, = 100, -0.8, 0.8


h_old = df.Histo1D(("1Dhist", "Reco (ditrack-Nbody).mass", nbins, xlow, xhigh), "Diff_meson_ditrack_mass")
h_mass = df.Histo1D(("1Dhist", "GEN (ditrack-Nbody).mass", nbins, xlow, xhigh), "Diff_meson_ditrack_mass_GEN")

h_old.SetLineColor(ROOT.kBlue)
h_mass.SetLineColor(ROOT.kRed)
canvas.cd(1)
h_old.Draw("hist")
canvas.cd(2)
h_mass.Draw("hist")

h_old.Scale(1/h_old.GetEntries())
h_mass.Scale(1/h_mass.GetEntries())

canvas.cd(3)
stack = ROOT.THStack("stack", "reco-gen")
stack.Add(h_old.GetValue())
stack.Add(h_mass.GetValue())
stack.Draw("hist nostack")

#stack.GetXaxis().SetRangeUser(100, 150)

legend = ROOT.TLegend(0.70, 0.70, 0.9, 0.9)
legend.SetMargin(0.17)
#legend.SetBorderSize(0)
legend.SetTextSize(0.04)
legend.AddEntry(h_old.GetValue(), "Reco", "l")
legend.AddEntry(h_mass.GetValue(), "Gen", "l")
legend.Draw()

canvas.SetGrid()
canvas.Draw()
```
